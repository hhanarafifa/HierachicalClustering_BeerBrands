```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
import scipy.cluster.hierarchy as sch # to create dendogram
from sklearn.cluster import AgglomerativeClustering # To create hierarchical clustering
```

```python
df = pd.read_csv('beer_data.csv')
# df
df_norm = norm_func(df.iloc[:,1:]) #eliminate unnecessary column
# df_norm
```

```python
# Do the normalization
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Create dendogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='complete'))
```

```python
# create 3 cluster
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'complete')
```

```python
y_hc = hc.fit_predict(df_norm) 
#the difference between fit_predict and fit is you dont have to assign labels_ attribute after fit()
df['clusterid'] = y_hc
df.sort_values('clusterid')
```
